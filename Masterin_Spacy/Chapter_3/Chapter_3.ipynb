### POS tagging 
POS tagging is the one of the most important linguistic features in the spacy, The POS is a part of syntantic category. For example: english has nine main categories verb, adverb, noun, pronoun, adjective, conjuction, determiner, intersection, and preposition. The set of categories, without any morpholoical and syntatic features are called **universeral tags**, spaCy uses the universal tags for POS. The POS is a statistical model because the tag of the current word depend on the neighbor word and the word tag. The POS was done by two main models, HMM and LSTM. Now a days all are using the LSTM for better accuracy, spacy also uses the same LSTM. 
 
 
**Why we need to learn the POS?**
The answer is simple: many
applications need to know the word type for better accuracy. Consider
machine translation systems for an example of this: the words for fish (V)
and fish (N) correspond to different words in Spanish, as illustrated in the
following code snippet:

I will fish/VB tomorrow. -> Pescaré/V mañana.

I eat fish/NN. -> Como pescado/N

Syntactic information can be used in many NLU tasks, and playing some POS
tricks can help your NLU code a lot

### Verb Tense and NLU

In the previous chapter, we used the example of the travel agency application
where we got the base forms (which are freed from verb tense and aspect) of
the verbs by using lemmatization. In this subsection, we'll focus on how to
use the verb tense and aspect information that we lost during the
lemmatization process.


**aspect -> (அம்சம்)** 

Verb tense and aspect are maybe the most interesting information that verbs
provide us, telling us when the action happened in time and if the action of
the verb is finished or ongoing. s. For instance, in the sentence I'm eating, the action eat
happens in the present and is ongoing, hence we describe this verb as present
progressive/continuous.


```Python
import spacy 

nlp = spacy.load("en_core_web_md") 

sent1 = "I flew to Rome."
sent2 = "I'm flying to Rome."
sent3 = "I will fly to Rome."

doc1 = nlp(sent1)
doc2 = nlp(sent2)
doc3 = nlp(sent3)
for doc in [doc1, doc2, doc3]:
    print([(w.text, w.lemma_) for w in doc if w.tag_== 'VBG' or w.tag_== 'VB'])
    
# VBG (a verb in present progressive form) eX: I'm eating (This referring the present, and it also continuing) 
# VB (a verb in base/ infinite form) 
```

````python
import spacy 

nlp = spacy.load("en_core_web_md") 

sent1 = "I flew to Rome."
sent2 = "I'm flying to Rome."
sent3 = "I will fly to Rome."

doc1 = nlp(sent1)
doc2 = nlp(sent2)
doc3 = nlp(sent3)
for doc in [doc1, doc2, doc3]:
    print([(w.text, w.lemma_) for w in doc if w.tag_== 'VBG' or w.tag_== 'VB'])
````

We iterated three doc objects one by one, and for each sentence we checked
if the fine-grained tag of the token is VBG (a verb in present progressive
form) or VB (a verb in base/infinitive form). Basically, we filtered out the
present progressive and infinitive verbs. You can think of this process as a
semantic representation of the verb in the form of (word form, lemma, tag)

### Dependency Parsing! 
In the previous topic, we discussed about the Pos- syntactic(தொடரியல்) categories of words. The POS tags provide information about neighbour's tags as well, they do not give away any relations between words that are not neighbours in the given sentence. The Dependency parsing is used to **analyze** sentence structures via dependencies between the tokens. **A dependency parser tags syntactic relations between tokens of the sentence and connects syntactically releated pairs of tokens, dependency or a dependency relation is a directed link between two tokens**. 

Dependency parser are just a tree like structure. 

<img src="images/tree.png" width="500"/>

All the sentence are a sequence of tokens. Each sentence are having te own linguistic structure. Every word in a sentence has a well defined role: such as verb, subject, object and so on; This kind of sentence structure is extensively used in the chat bots and machine translation. 

Spacy assign each token a dependency label, just as other linguistic features such as lemma_ or POS_ tag. Spacy shows dependecy relations between  the lables.

<img src="images/amod.png" width="500"/>

**Amod** (adjective-noun relation)

**The dependecy label describes a type of syntactic relation between two tokens**. One of the token is **Syntactic parent** and other is **syntactic child**. The dependency label is assingned to the child. To get the dependency we have two snips **dep(int)** and **dep_(unicode)**. 

```Python 
doc = nlp('Blue Flower') 

for token in doc: 
    print(token.text, token.dep_)  
    
# outputs: 

Blue compound
Flower ROOT
```
ROOT is a special label in the dependency parser. it will assign to the main verb of the sentence. Same like if we are proceeding the phrase(not a full sentence), the root word is assign to the phrase. Same we have phrase "Blue Flower" the root word is assign to the flower. All the sentences / phrase have only one root. The tree structure have more than one child but it has only one parent node(root node). 

````python
doc = nlp('Blue Flower') 

for token in doc: 
    print(token.text, token.dep_)  
````

The pronoun we is labeled by nsubj, hence

So far, so good. Now, what is happening on the right side? trying is attached
to the second verb understand via an xcomp relation. The xcomp (or open
complement) relation of a verb is a clause without its own subject. Here, the

complement. We follow the dobj arc from the second verb, understand, and
land on the noun, difference, which is the direct object of the to understand
the difference clause

### NER 

**What is a named entity?** A named entity is a real-world object that we can
refer to by a proper name or a quantity of interest. It can be a person, a place
(city, country, landmark, famous building), an organization, a company, a
product, dates, times, percentages, monetary amounts, a drug, or a disease
name. Some examples are Alicia Keys, Paris, France, Brandenburg Gate,
WHO, Google, Porsche Cayenne, and so on.

A named entity always points to a specific object, and that object is
distinguishable via the corresponding named entity. For instance, if we tag
the sentence Paris is the capital of France, we parse Paris and France as
named entities, but not the word capital. The reason is that capital does not
point to a specific object; it's a general name for many objects.

````python
from bs4 import BeautifulSoup
import requests
import spacy

def url_text(url_string):
    res = requests.get(url_string)
    html = res.text
    soup = BeautifulSoup(html, 'lxml')
    for script in soup(["script", "style", 'aside']):
        script.extract()
    text = soup.get_text()
    return " ".join(text.split())


ny_art = url_text("https://www.nytimes.com/2021/01/12/opinion/trump-america-allies.html")
nlp = spacy.load("en_core_web_md")
doc = nlp(ny_art)
````

BeautifulSoup is a popular Python package for extracting text from HTML
and Extensible Markup Language (XML) documents. Then, we created a
nlp object, passed the article body to the nlp object, and created a Doc
object.

### Merging and splitting Tokens 

If we want to join or split the NER, we can use the retokenizer(), it helps to join and split the tokens

**Merging the tokens**

```Python 
doc = nlp("She lived in New Hampshire.")

ind = [(token.text, token.i) for token in doc] 
print(Fore.MAGENTA+str(ind))

with doc.retokenize() as retokenizer: 
    retokenizer.merge(doc[3:5], attrs= {"LEMMA": "new hampshire"}) 
    # attrs is a dictionary of token attributes we want to assign to the new token, such as lemma, pos, tag, ent_type, and so on.
   
print(Fore.MAGENTA+ str([(token.text, token.i) for token in doc]))

print(Fore.CYAN+ str( [(token.lemma_) for token in doc]) ) 

# output 

(New Hampshire,)
[('She', 0), ('lived', 1), ('in', 2), ('New', 3), ('Hampshire', 4), ('.', 5)]  # This is before the split 
[('She', 0), ('lived', 1), ('in', 2), ('New Hampshire', 3), ('.', 4)]  # This is after the split 
['she', 'live', 'in', 'new hampshire', '.']  # now our entire structure is changed !

```

````python
# Mergeing the tokens 

doc = nlp("She lived in New Hampshire.")
print(doc.ents) 
print(Fore.BLUE+str('Before splitting')) 
ind = [(token.text, token.i) for token in doc] 
print(Fore.MAGENTA+str(ind))

with doc.retokenize() as retokenizer: 
    retokenizer.merge(doc[3:5], attrs= {"LEMMA": "new hampshire"})
   
print(Fore.GREEN+'After Splitting')
print(Fore.MAGENTA+ str([(token.text, token.i) for token in doc]))

print()
print(Fore.CYAN+ str( [(token.lemma_) for token in doc]) ) 

````

Now we are going to split the **New Hampshire** to **New** (individual token) and **Hampshire** (individual token), to split this we need to add head parameter inside. To change the head parameter we need to look on two main considerations. 

Firstly, if you give a relative position (such as (doc[3], 1)) in the
following code segment, this means that head of doc[3] will be the +1th
position token—that is, doc[4] in the new setup (please see the following
visualization ).

Secondly, if you give an absolute position, it means the position in the
original Doc object. In the following code snippet, the second item in the
heads list means that the Hampshire token's head is the second token in
the original Doc, which is the in token


````python
(doc[3], 1)  # relative position 
````

````python
doc[2]  # absolute position 
````

````python
with doc.retokenize() as retokenizer:
    heads = [(doc[3], 1), doc[2]]
    attrs = {"TAG":["NNP", "NNP"],"DEP": ["compound", "pobj"]}
    
    retokenizer.split(doc[3], ["New", "Hampshire"], heads=heads, attrs=attrs)
print([(token.text, token.lemma_, token.i) for token in doc])
````

````python
print([(token.text, token.lemma_, token.i) for token in doc])

for token in doc:
    print(f"{Fore.RED+'Text'}: {Fore.GREEN +token.text}, {Fore.RED+'POS'}: {Fore.GREEN + token.pos_}, {Fore.RED +'TAG'}: {Fore.GREEN + token.tag_}, {Fore.RED+'DEP'}:{Fore.GREEN + token.dep_}")

displacy.render(doc, style = 'dep')
````
